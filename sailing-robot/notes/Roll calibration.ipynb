```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def add_pitch_roll(df):
    df['pitch'] = np.degrees(np.arctan2(df.acc_x, np.sqrt(df.acc_y**2 + df.acc_z**2)))
    df['roll'] = np.degrees(np.arctan2(-df.acc_y, -df.acc_z))
```

```python
rock_boldrewood = pd.read_csv('../recorded_data/calibration_roll_boldrewood_2017-07-04T18.17.32.csv')
add_pitch_roll(rock_boldrewood)
```

```python
def plot_calib_roll(df):
    fig, (ax0, ax1) = plt.subplots(ncols=2, figsize=(14, 4))
    df.plot(y='mag_x', ax=ax0)
    df.plot(y='mag_y', ax=ax0)
    df.plot(y='mag_z', ax=ax0)

    df.plot(y='pitch', ax=ax1)
    df.plot(y='roll', ax=ax1)
    return fig, (ax0, ax1)
```

```python
fig, axes = plot_calib_roll(rock_boldrewood)
fig.suptitle('Outsite B176, 2017-07-04')
```

```python
rock_eastleigh = pd.read_csv('../recorded_data/calibration_roll_eastleigh_2017-06-17.csv')
add_pitch_roll(rock_eastleigh)
fig, axes = plot_calib_roll(rock_eastleigh)
fig.suptitle('Eastleigh lake, 2017-06-17')
```
